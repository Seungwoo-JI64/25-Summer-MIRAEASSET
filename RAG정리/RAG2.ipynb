- 유사도 올릴방법

```python
import os
import pandas as pd
import numpy as np
from dotenv import load_dotenv
from supabase import create_client, Client
from sklearn.metrics.pairwise import cosine_similarity
import ast

load_dotenv()

url = os.getenv("SUPABASE_URL")
key = os.getenv("SUPABASE_KEY")

supabase: Client = create_client(url, key)
```

```python
df_company = pd.DataFrame(supabase.table("company_summary").select("company_name,summary, summary_embedding").execute().data)
df_company
```

```python
df_news = pd.DataFrame(supabase.table("financial_news_summary").select("title, url, summary, embedding").execute().data)
df_news
```

```python
#차원확인
import ast
import numpy as np

# 문자열 형태의 벡터를 실제 벡터로 변환
df_company['embedding_array'] = df_company['summary_embedding'].apply(lambda x: np.array(ast.literal_eval(x)))

# 각 임베딩의 차원 확인
dimensions = df_company['embedding_array'].apply(lambda x: x.shape[0]).unique()
print("임베딩 차원들:", dimensions)
```

```python
df_news['embedding_array'] = df_news['embedding'].apply(lambda x: np.array(ast.literal_eval(x)))

# 각 임베딩의 차원 확인
dimensions = df_news['embedding_array'].apply(lambda x: x.shape[0]).unique()
print("임베딩 차원들:", dimensions)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import ast

# 1. 문자열 → 벡터 변환
df_company['embedding_array'] = df_company['summary_embedding'].apply(lambda x: np.array(ast.literal_eval(x)))
df_news['embedding_array'] = df_news['embedding'].apply(lambda x: np.array(ast.literal_eval(x)))

# 2. 결과 저장용 딕셔너리
top_news_per_company = {}

# 3. 각 기업마다 유사도 계산
for idx, row in df_company.iterrows():
    company_name = row['company_name']
    company_vec = row['embedding_array'].reshape(1, -1)  # (1, 768)

    # 뉴스 전체 임베딩과의 유사도 계산
    news_embeddings = np.vstack(df_news['embedding_array'].values)
    similarities = cosine_similarity(company_vec, news_embeddings)[0]  # shape: (n_news, )

    # 유사도 상위 15개 뉴스 인덱스 추출
    top_indices = similarities.argsort()[-15:][::-1]

    # 해당 뉴스 추출 및 유사도 함께 저장
    top_news = df_news.iloc[top_indices][['title', 'summary', 'url']].copy()
    top_news['similarity'] = similarities[top_indices]

    top_news_per_company[company_name] = top_news

# 4. 예시 출력 (삼성전자 관련 뉴스 Top 15)
print("📌 삼성전자 관련 뉴스 Top 15:")
print(top_news_per_company['Samsung Biologics'])
```

```python
# dict의 각 DataFrame에 company_name 컬럼 추가 후 통합
company_news_15 = pd.concat([
    df.assign(company_name=company) 
    for company, df in top_news_per_company.items()
], ignore_index=True)
company_news_15 = company_news_15[['company_name', 'title', 'summary', 'url', 'similarity']]
# 예시 출력
company_news_15
```

##### supabase에 업로드

```python
company_news_15 = company_news_15.reset_index(drop=True)
company_news_15['id'] = company_news_15.index + 1
```

```python
# DataFrame의 각 row를 dict로 변환
records = company_news_15.to_dict(orient='records')
# 예: 테이블 이름이 'company_news_15'
response = supabase.table("company_news_15").insert(records).execute()

# 결과 확인
print(response)
```
