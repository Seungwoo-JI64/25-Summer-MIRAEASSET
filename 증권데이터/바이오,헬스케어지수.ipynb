2. 미국 FDA 신약 허가 일정

```python
import requests
import pandas as pd

url = "https://api.fda.gov/drug/drugsfda.json"
params = {'limit': 5}

response = requests.get(url, params=params)
data = response.json()

approval_list = []
for item in data.get('results', []):
    app_num = item.get('application_number')
    sponsor = item.get('sponsor_name')
    for product in item.get('products', []):
        brand_name = product.get('brand_name')
        
        # submission 배열 안에서 승인일 찾기
        approval_date = None
        submissions = product.get('submission', [])
        for sub in submissions:
            # 승인일이 있으면 그걸로 설정
            if 'approval_date' in sub:
                approval_date = sub['approval_date']
                break  # 가장 첫 승인일만 사용
        
        approval_list.append({
            '신약명': brand_name,
            '승인일': approval_date,
            '신청번호': app_num,
            '제약사': sponsor
        })

df = pd.DataFrame(approval_list)
print(df)
```

-> 승인일이 다 none으로 나오고, 신약 승인 목표일 같은 경우에는 직접 계속 찾아보면서 크롤링을 해야해서 하기 좀 많이 어렵습니다.
