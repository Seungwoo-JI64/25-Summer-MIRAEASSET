```python
import yfinance as yf

# FDN ETF 데이터 60일치 다운로드
data = yf.download('FDN', period='60d')

# 날짜 인덱스를 컬럼으로 변경하고, 날짜와 종가 컬럼만 선택
fdn = data.reset_index()[['Date', 'Close']]

# 컬럼명 한글로 변경
fdn.columns = ['날짜', '종가']

fdn
```

->다우존스 관련해서 하려했는데 api에서 제공을 제대로 안 해줍니다. 따라서 다우존스 인터넷 컴포지트 지수를 추종하는 etf로 변경하였으며, 이게 더 거래량이 많다고 하네요.
