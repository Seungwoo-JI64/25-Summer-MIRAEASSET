```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

ticker = 'LIT'  # 리튬 ETF

raw = yf.download(ticker, period='90d', interval='1d')
raw = raw.reset_index()[['Date', 'Close']]
raw.columns = ['날짜', 'LIT가격']

cutoff = datetime.today() - timedelta(days=60)
lit_price = raw[raw['날짜'] >= cutoff].reset_index(drop=True)
lit_price
```
