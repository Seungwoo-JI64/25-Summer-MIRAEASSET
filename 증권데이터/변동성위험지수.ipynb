```python
import yfinance as yf
import pandas as pd

# CBOE 변동성 지수(VIX) 다운로드
CBOE_raw= yf.download('^VIX', period='60d')

# 날짜를 인덱스에서 컬럼으로, 종가만 추출
CBOE = CBOE_raw[['Close']].reset_index()

# 컬럼명 한글로 변경
CBOE.columns = ['날짜', '종가']

# 결과 확인
CBOE
```
