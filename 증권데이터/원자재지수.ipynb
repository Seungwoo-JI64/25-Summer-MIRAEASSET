```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# WTI 원유 선물 티커
ticker = 'CL=F'

# 90일치 데이터 수집
raw = yf.download(ticker, period='90d', interval='1d')

# 날짜 및 종가만 추출 후 필터링
TWI = raw.reset_index()[['Date', 'Close']]
TWI.columns = ['날짜', '가격']

# 오늘 기준 60일 이내 데이터만 남기기
cutoff_date = datetime.today() - timedelta(days=60)
TWI = TWI[TWI['날짜'] >= cutoff_date].reset_index(drop=True)
TWI
```
