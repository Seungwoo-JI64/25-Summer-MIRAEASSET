```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# 환율 티커 (USD/KRW)
ticker = 'USDKRW=X'

# 90일치 데이터 받아온 뒤, 그중 최근 60일 내로 필터링
raw = yf.download(ticker, period='90d', interval='1d')

# 인덱스를 날짜 컬럼으로 변환
fx = raw.reset_index()[['Date', 'Close']]
fx.columns = ['날짜', '환율']

# 오늘 기준 60일 전 날짜 계산
cutoff_date = datetime.today() - timedelta(days=60)

# 60일 이내 데이터만 필터링
fx = fx[fx['날짜'] >= cutoff_date]

# 인덱스 초기화
fx = fx.reset_index(drop=True)

fx
```

-> 여기서 주말같은 공휴일은 환율 데이터가 없어 60일을 그냥 코드에 넣으면 더 먼 과거인 4월 데이터도 나옵니다. 따라서 코드를 현재를 기준으로  60일까지만 나오게 짰어요
