```python
import yfinance as yf
import pandas as pd

# 티커 설정 (S&P 500 지수)
ticker = '^GSPC'

# 최근 60일치 데이터 다운로드
snp500 = yf.download(ticker, period='60d')

# 종가 데이터만 추출
close_prices = snp500['Close']

snp500 = pd.DataFrame(snp500['Close'])
snp500.reset_index(inplace=True)  # 날짜 컬럼을 일반 컬럼으로 변환
snp500.columns = ['날짜', '종가']

snp500
```

2. 나스닥 100

```python
nasdaq100 = yf.download('^NDX', period='60d')

# 종가만 추출해서 데이터프레임 만들기
nasdaq100 = pd.DataFrame(nasdaq100['Close'])
nasdaq100.reset_index(inplace=True)  # 날짜를 컬럼으로 변환
nasdaq100.columns = ['날짜', '종가']

nasdaq100
```
